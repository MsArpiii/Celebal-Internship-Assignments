**Your Geometric Explanation (4.3):**

Geometrically, a matrix transformation can rotate, stretch, compress, or even flip vectors. Eigenvectors are special vectors whose direction remains unchanged after the transformation—they continue to lie along the same line, although their length may change. The corresponding eigenvalues indicate how much the matrix stretches or compresses these vectors. ### 4.4 SVD & Dimensionality Reduction

1. Create a random 4×3 matrix `X` (use `np.random.seed(42)`).

4. Perform a **rank-1 approximation** using only the largest singular value/vector — call it `X_approx`.
5. Explain the relationship between SVD components and PCA in the markdown cell below.

````python
np.random.seed(42)
X = np.random.randn(4, 3)

U, S, Vt = None, None, None  # YOUR CODE HERE
U, S, Vt = np.linalg.svd(
    X, full_matrices=False
)

X_reconstructed = None  # YOUR CODE HERE — reconstruct from U, S, Vt
X_reconstructed = U @ np.diag(S) @ Vt

X_approx        = None  # YOUR CODE HERE — rank-1 approximation
X_approx        = np.outer(
    U[:, 0],
    Vt[0, :]
) * S[0]

print(f"U shape: {U.shape}, S shape: {S.shape}, Vt shape: {Vt.shape}")
print(f"Singular values: {S}")
print(f"Reconstruction error: {np.linalg.norm(X - X_reconstructed):.2e}")
print(f"Rank-1 approximation:\n{X_approx}")

assert np.allclose(X, X_reconstructed, atol=1e-10), "Reconstruction failed!"
print("4.4 passed ✓")
````

**SVD → PCA Connection (4.4):**

> *Explain: what are U, S, Vt in SVD, and how do they relate to principal components? Which matrix contains the directions of maximum variance?*

In Singular Value Decomposition (SVD), U contains the left singular vectors, which describes patterns in the rows of the data, S contains the singular values, which indicates the importance or amount of variance captured by each component, and Vt contains the right singular vectors. In Principal Component Analysis (PCA), the principal components correspond to the right singular vectors in Vt (or the columns of V), and the singular values indicate how much variance each component explains. Therefore, the directions of maximum variance are contained in Vt, with the first singular vector representing the direction of the greatest variance in the data.

(Also
U = left singular vectors
S = singular values
Vt = transpose of V
V = right singular vectors)

### 5.5 Model Monitoring Concepts

1. Implement `compute_psi(expected, actual, bins=10)` that computes the **Population Stability Index** between two distributions.
2. Simulate a scenario where a model's input distribution shifts and plot both distributions.
3. In the markdown cell, define: *concept drift*, *covariate drift*, PSI thresholds (what values indicate no/minor/major shift), and one trigger for retraining.

````python
def compute_psi(expected, actual, bins=10):
    """
    PSI = sum((actual_pct - expected_pct) * ln(actual_pct / expected_pct))
    Add epsilon=1e-10 to avoid log(0).
    """
    # YOUR CODE HERE
    epsilon = 1e-10

    breaks = np.linspace(
        min(expected.min(), actual.min()),
        max(expected.max(), actual.max()),
        bins + 1
    )

    expected_counts, _ = np.histogram(
        expected,
        bins=breaks
    )

    actual_counts, _ = np.histogram(
        actual,
        bins=breaks
    )

    expected_pct = expected_counts / len(expected)
    actual_pct = actual_counts / len(actual)

    psi = np.sum(
        (actual_pct - expected_pct)
        *
        np.log(
            (actual_pct + epsilon)
            /
            (expected_pct + epsilon)
        )
    )

    return psi
    pass

np.random.seed(1)
train_dist = np.random.normal(50, 10, 1000)    # training distribution
drift_dist = np.random.normal(65, 12, 1000)    # shifted production distribution

psi_value = compute_psi(train_dist, drift_dist)
print(f"PSI: {psi_value:.4f}")
print(f"Shift severity: {'Major' if psi_value > 0.2 else 'Minor' if psi_value > 0.1 else 'Stable'}")

# Plot both distributions
fig, ax = plt.subplots()
# YOUR CODE HERE
train_counts, bins = np.histogram(
    train_dist,
    bins=20,
    density=True
)

drift_counts, _ = np.histogram(
    drift_dist,
    bins=bins,
    density=True
)

ax.fill_between(
    bins[:-1],
    train_counts,
    alpha=0.5,
    label='Train'
)

ax.fill_between(
    bins[:-1],
    drift_counts,
    alpha=0.5,
    label='Drift'
)

ax.legend(); ax.set_title('Distribution Shift Visualization')
plt.tight_layout(); plt.show()
````

**Concept Drift vs Covariate Drift & PSI Thresholds (5.5):**

- **Concept drift:** It occurs when the relationship between input features and the target variable changes over time. For example, factors that previously influenced employee performance may become less relevant in the future.
- **Covariate drift:** It occurs when the distribution of input features changes, even though the relationship with the target remains the same. For example, the salary distribution of employees may shift over time due to hiring trends or market conditions.

**PSI Interpretation**

**Population Stability Index (PSI)** measures how much a data distribution has changed compared to a reference distribution.
- **PSI < 0.1:** Stable distribution (little or no drift)
- **PSI 0.1–0.2:** Moderate drift, should be monitored
- **PSI > 0.2:** Significant drift, model retraining may be required

PSI helps identify data shifts early and ensures that machine learning models continue to perform reliably over time.
- **Retraining trigger example:** If the PSI remains above 0.2 for multiple monitoring periods, it may indicate significant data drift and the model should be retrained using more recent data.

### 6.3 Bayes' Theorem

**Scenario — Spam filter:**
- P(Spam) = 0.30 (prior)
- P("free" | Spam) = 0.80 (likelihood)
- P("free" | Not Spam) = 0.05

1. Compute P(Spam | "free") using Bayes' theorem.
2. Implement a simple `naive_bayes_predict(prior_spam, p_word_given_spam, p_word_given_ham)` that returns the posterior P(Spam | word).
3. In the markdown cell, map each term — **prior, likelihood, evidence, posterior** — to the spam example.

````python
p_spam            = 0.30
p_free_given_spam = 0.80
p_free_given_ham  = 0.05
p_ham             = 1 - p_spam

# P(free) — law of total probability
p_free = None  # YOUR CODE HERE
p_free = (
    p_free_given_spam*p_spam
    +
    p_free_given_ham*p_ham
)

# Bayes
p_spam_given_free = None  # YOUR CODE HERE
p_spam_given_free = (
    p_free_given_spam*p_spam
) / p_free

print(f"P('free'): {p_free:.4f}")
print(f"P(Spam | 'free'): {p_spam_given_free:.4f}")

def naive_bayes_predict(prior_spam, p_word_given_spam, p_word_given_ham):
    # YOUR CODE HERE
    p_ham = 1 - prior_spam

    evidence = (
        p_word_given_spam*prior_spam
        +
        p_word_given_ham*p_ham
    )

    return (
        p_word_given_spam*prior_spam
    ) / evidence

    pass

pred = naive_bayes_predict(0.30, 0.80, 0.05)
assert abs(pred - p_spam_given_free) < 1e-9
print(f"naive_bayes_predict: {pred:.4f}")
print("6.3 passed ✓")
````

### 6.4 Central Limit Theorem

1. Define a population following an **exponential distribution** (λ=1, size=100,000).
2. Repeatedly draw samples of size `n=30` and record their means — do this 5,000 times.
3. Plot the histogram of sample means.
4. Overlay the **theoretical normal distribution** predicted by the CLT: μ = population mean, σ = population_std / sqrt(n).
5. Run a KS test to confirm the distribution of sample means is approximately normal.

````python
np.random.seed(7)
population = np.random.exponential(scale=1.0, size=100_000)

n_samples = 5000
sample_size = 30
sample_means = None  # YOUR CODE HERE — array of 5000 sample means
sample_means = np.array([
    np.mean(
        np.random.choice(
            population,
            sample_size
        )
    )
    for _ in range(n_samples)
])

pop_mean = population.mean()
pop_std  = population.std()
clt_std  = None  # YOUR CODE HERE — predicted std of sample means
clt_std = pop_std / np.sqrt(sample_size)

print(f"Population mean: {pop_mean:.4f}, Population std: {pop_std:.4f}")
print(f"Sample means mean: {sample_means.mean():.4f}, std: {sample_means.std():.4f}")
print(f"CLT predicted std: {clt_std:.4f}")

# Plot
fig, ax = plt.subplots()
# YOUR CODE HERE — histogram of sample means + theoretical normal overlay
ax.hist(
    sample_means,
    bins=30,
    density=True,
    alpha=0.6
)

x = np.linspace(
    sample_means.min(),
    sample_means.max(),
    300
)

ax.plot(
    x,
    stats.norm.pdf(
        x,
        pop_mean,
        clt_std
    )
)

ax.set_title('CLT: Distribution of Sample Means')
plt.tight_layout(); plt.show()

# KS test
ks_result = stats.kstest(sample_means, 'norm', args=(pop_mean, clt_std))
print(f"KS test p-value: {ks_result.pvalue:.4f} → Approximately normal? {ks_result.pvalue > 0.05}")
````

**CLT Reflection (6.4):**

> *In 2–3 sentences: Why does this result matter for statistics and machine learning? What assumption does the CLT justify in many classical statistical tests?*

The **Central Limit Theorem (CLT)** states that the distribution of sample means becomes approximately normal as the sample size increases, regardless of the shape of the original population. In this experiment, although the population may not be normally distributed, the histogram of sample means forms a bell-shaped curve that aligns closely with the normal curve. This is important because it enables statistical inference, confidence intervals, and hypothesis testing, which are widely used in data analysis and machine learning.
